## Preparing the data

In this section, we will be preparing the data by beginning to combine all of the datasets into one pandas dataframe and do a bit of data cleaning. Deleting unnecessary 
columns and removing null values.

```python
del price_data["month"]
price_data.columns = ["price", "value"]
```

## Creating a target

We will be creating a target that we will be trying to predict with our model. Our target will be to predict if the prices will go up or down after 13 days. If the price goes up our `change` variable will be `1`, and if it goes down, the value will be `0`

Next, we'll shift the day from previous days "forward" 13 days, so we can use it to predict our target

Then we'll combine both so we have a set of training data


```python
price_data["adj_price"] = price_data["price"] / price_data["cpi"] * 100
price_data["adj_value"] = price_data["value"] / price_data["cpi"] * 100
```

```python
price_data.plot.line(y="price", use_index=True)
```

```python
price_data.plot.line(y="adj_price", use_index=True)
```

```python
price_data["next_quarter"] = price_data["adj_price"].shift(-13)
```

```python
price_data.dropna(inplace=True)
```

```python
price_data
```

```python
price_data["change"] = (price_data["next_quarter"] > price_data["adj_price"]).astype(int)
```

```python
price_data
```

```python
price_data["change"].value_counts()
```

```python
predictors = ["interest", "vacancy", "adj_price", "adj_value"]
target = "change"
```

## Creating a machine learning model

Next, we'll create a machine learning model to see how accurately we can predict the trends of prices. We will implement a random forest classifier to generate our predictions.

To measure the accuracy of our predictions, we will be using the error metric accuracy score from `scikit-learn`

We will also be creating a backtesting engine to improve our predictions. Backtesting ensures that we only use data from before the day that we're predicting. If we use data from after the day we're predicting, the algorithm will not be applicable to the real world.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np

START = 260
STEP = 52

def predict(train, test, predictors, target):
    rf = RandomForestClassifier(n_estimators=200,min_samples_split=10, random_state=1)
    rf.fit(train[predictors], train[target])
    preds = rf.predict(test[predictors])
    return preds

def backtest(data, predictors, target):
    all_preds = []
    for i in range(START, data.shape[0], STEP):
        train = price_data.iloc[:i]
        test = price_data.iloc[i:(i+STEP)]
        all_preds.append(predict(train, test, predictors, target))
    
    preds = np.concatenate(all_preds)
    return preds, accuracy_score(data.iloc[START:][target], preds)
```

```python
preds, accuracy = backtest(price_data, predictors, target)
```

```python
preds
```

```python
accuracy
```

## Improving accuracy

We will try using different predictors to improve the accuracy of our model

We will create a predictor for yearly ratios, which is the yearly rolling average.

```python
yearly = price_data.rolling(52, min_periods=1).mean()
```

```python
yearly_ratios = [p + "_year" for p in predictors]
price_data[yearly_ratios] = price_data[predictors] / yearly[predictors]
```

```python
price_data
```

```python
preds, accuracy = backtest(price_data, predictors + yearly_ratios, target)
```

```python
accuracy
```

```python
pred_match = (preds == price_data[target].iloc[START:])
pred_match[pred_match == True] = "green"
pred_match[pred_match == False] = "red"
```

```python
import matplotlib.pyplot as plt

plot_data = price_data.iloc[START:].copy()

plot_data.reset_index().plot.scatter(x="index", y="adj_price", color=pred_match)
```

```python
from sklearn.inspection import permutation_importance

rf = RandomForestClassifier(min_samples_split=10, random_state=1)
rf.fit(price_data[predictors], price_data[target])

result = permutation_importance(rf, price_data[predictors], price_data[target], n_repeats=10, random_state=1)
```
